```python
data = df.T
data
```

```python

columns = list(data)
for col in columns[0:]: 
    data[col] = np.where(data[col] >= 0.5, 1, data[col])
    data[col] = np.where(data[col] < 0.5, 0, data[col])


```

```python
data
```

# APRIORI

```python
import mlxtend
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
frequent_itemsets = apriori(data, min_support=0.5, use_colnames=True, low_memory=True)
```

```python
frequent_itemsets
```

```python
frequent_itemsets['length'] = frequent_itemsets['itemsets'].apply(lambda x: len(x))
frequent_itemsets
```

```python
frequent_itemsets[ (frequent_itemsets['length'] >= 2) &
                   (frequent_itemsets['support'] >= 0.5) ]
```

```python
rules=association_rules(frequent_itemsets, metric="confidence", min_threshold=0.2)
```

```python
rules["antecedent_len"] = rules["antecedents"].apply(lambda x: len(x))
rules
```

```python
rules[ (rules['antecedent_len'] >= 1) &
       (rules['confidence'] > 0.20)&
       (rules['support'] > 0.05)]
```
