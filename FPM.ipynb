# APRIORI

```python
import mlxtend
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
import timeit
```

```python
start = timeit.default_timer()
apriori_frequent_itemsets = apriori(min_max_data, min_support=0.4, use_colnames=True)
stop = timeit.default_timer()
print (stop-start)
```

```python
#apriori_frequent_itemsets
```

```python
apriori_rules=association_rules(apriori_frequent_itemsets, metric="confidence", min_threshold=0.5)
```

```python
apriori_rules["antecedent_len"] = apriori_rules["antecedents"].apply(lambda x: len(x))
apriori_rules
```

```python
apriori_rules[ (apriori_rules['antecedent_len'] >= 1) &
       (apriori_rules['confidence'] > 0.6)&
       (apriori_rules['support'] > 0.5)&
       (apriori_rules['lift'] > 1.2)]
```

# ECLAT

```python
start = timeit.default_timer()
eclat_frequent_itemsets = apriori(min_max_data, min_support=0.4, use_colnames=True)
stop = timeit.default_timer()
print (stop-start)
```

```python
eclat_rules=association_rules(eclat_frequent_itemsets, metric="support", min_threshold=0.5)
```

```python
eclat_rules["antecedent_len"] = eclat_rules["antecedents"].apply(lambda x: len(x))
eclat_rules
```

# FP-GROWTH

```python
from mlxtend.frequent_patterns import fpgrowth
```

```python
start = timeit.default_timer()
fp_frequent_itemsets=fpgrowth(min_max_data, min_support=0.4, use_colnames=True)
stop = timeit.default_timer()
print (stop-start)
```

```python
rules = association_rules(fp_frequent_itemsets, metric="confidence", min_threshold=0.5)
```

```python
rules["antecedent_len"] = rules["antecedents"].apply(lambda x: len(x))
rules
```
